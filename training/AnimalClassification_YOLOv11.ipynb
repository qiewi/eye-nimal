```python
from ultralytics import YOLO

# Pilih model klasifikasi YOLOv8 (misal, small)
model = YOLO('yolo11l-cls.pt')  # atau 'yolov8n-cls.pt', 'yolov8m-cls.pt', dst

# Training
results = model.train(
    data='/kaggle/input/yoloanimalclassification',  # path ke folder dataset
    epochs=20,
    imgsz=224,                  # ukuran gambar, default 224 untuk klasifikasi
    batch=64,
    project='yolo-animals10-cls-exp',
    name='yolo11l-cls',
    device=0,                    # atau 'cpu'
    seed = 42,
    val=True,
    lr0 = 1e-4,
    weight_decay=1e-5,
    save_json = True,
    save_txt = True,
    patience = 3,
    optimizer = 'AdamW'
)
```

```python
# Validate the model
metrics = model.val(save_json=True)  # no arguments needed, dataset and settings remembered
df = metrics.to_df()
df.to_csv('yolo_val_metrics.csv', index=False)
```
